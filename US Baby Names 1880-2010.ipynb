```python
import numpy as np
from pandas import DataFrame, Series
import pandas as pd
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
years = range(1880, 2011)
pieces = []
for year in years:
    piece = pd.read_csv(f'Data_Sets/babynames/yob{year}.txt', names = ['name', 'sex', 'births'])
    piece['year'] = year
    pieces.append(piece)
```

```python
pieces[0]
```

```python
df = pd.concat(pieces, ignore_index=True)
```

```python
df
```

```python
population_each_year = df.groupby('year').births.sum()
```

```python
population_each_year
```

```python
population_each_year.plot(style='c--',title='Total births by year')
```

```python
by_sex = df.pivot_table('births', columns='sex', index='year', aggfunc='sum')
by_sex
```

```python
by_sex.plot(title='Total births by sex and year')
```

```python
groups = df.groupby(['year','sex'])
```

```python
def add_prop(group):
    group['prop'] = group.births / group.births.sum()
    return group
df = groups.apply(add_prop)
```

```python
df
```

```python
pieces = []
for year, group in df.groupby(['year', 'sex']):
    pieces.append(group.sort_values(by='births', ascending=False)[:1000])
top1000 = pd.concat(pieces, ignore_index=True)
```
